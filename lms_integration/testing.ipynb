```python
from reportlab.pdfgen import canvas

# create a pdf

def create_nomral():
    #  create  a canvas
    c = canvas.Canvas("normal.pdf")
    #  add a text
    c.drawString(100, 100, "Hello, World!")
    #  save the pdf
    c.save()
    print("PDF created")

# create a blank pdf

def create_blank():
    c = canvas.Canvas("blank.pdf")
    c.showPage()     # just an empty page

    c.save()
    print("Blank PDF created")

# create a pdf with a table

# create a gibersh pdf

def create_gibersh():
    c = canvas.Canvas("gibersh.pdf")
    # giberish  text
    c.drawString(100, 100, "Hkajflakjfdla !")
    c.drawString(100, 200, "dlfjlakfjlakjflak !")   
    c.drawString(100, 300, "lsjldjaflkdfj !")
    c.drawString(100, 400, "aljflasjlfdjaldj !")
    c.drawString(100, 500, "=kslfjlaskfjl !")
    c.drawString(100, 600, "lsjjalfjadl !")
    c.drawString(100, 700, "adfjajflk !")
    c.drawString(100, 800, "aldfjlakfl !")
    c.save()
    print("Gibersh PDF created")


#  call all the fucntion 
create_nomral()
create_blank()
create_gibersh()


```
